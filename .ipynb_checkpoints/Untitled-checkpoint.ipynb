```python
import pandas as pd
import glob
import os

import matplotlib.pyplot as plt
```

## Carregando os arquivos

```python
print('\n\nLendo os arquivos CSV...')

colunas = ["Time since reaction (ms)", "Filter (adc)"]

base_dir = os.getcwd()
path = os.path.join(base_dir, 'Desafio_2_2026_1')

print("Caminho final:", path)

lotes = ['Lote 062506', 'Lote 072501']
solucoes = ['Nivel I', 'Nivel III', 'Solucao 0', 'Solucao 1', 'Solucao 2']

lista_df = []

for lote in lotes:
    for solucao in solucoes:

        pasta = os.path.join(path, lote, solucao)

        arquivos = glob.glob(os.path.join(pasta, '*.csv')) + \
                   glob.glob(os.path.join(pasta, '*.CSV'))

        print(f"{lote} - {solucao} -> {len(arquivos)} arquivos")

        for arquivo in arquivos:

            df = pd.read_csv(arquivo)
            df.columns = df.columns.str.strip()

            df = df[colunas]

            df["Lote"] = lote
            df["Solucao"] = solucao

            lista_df.append(df)

if not lista_df:
    raise ValueError("Nenhum CSV encontrado nas pastas.")

df = pd.concat(lista_df, ignore_index=True)

for col in colunas:
    df[col] = df[col].apply(lambda x: float(int(str(x), 16)))

print('\nDataset lido:', df.shape)
print('\nDataset processado:\n', df.head())

output_path = os.path.join(base_dir, 'dataset_processado.csv')
df.to_csv(output_path, index=False)

print('\nDataset salvo em dataset_processado.csv')
```

```python
df = df.rename(columns={"Time since reaction (ms)": "time", "Filter (adc)":"adc", "Solucao":"solucao", "Lote":"lote"})
```

# EDA

## Verificando dados nulos

```python
print("\nValores nulos:\n", df.isna().sum())
```

## Estatística Descritivas

```python
print('\nEstatistica Descritiva:\n\n',df.describe())
```

## Gráfico de Dispersão

```python
for solucao in df["solucao"].unique():

    subset = df[df["solucao"] == solucao]

    plt.scatter(
        subset["time"],
        subset["adc"],
        label=solucao,
        alpha=0.6,
        s=1
    )

plt.legend()
plt.grid(True)
plt.show()
```

```python


for lote in df["lote"].unique():

    plt.figure(figsize=(10,5))

    for solucao in df["solucao"].unique():

        subset = df[(df["lote"] == lote) & (df["solucao"] == solucao)]

        plt.scatter(
            subset["time"],
            subset["adc"],
            label=solucao,
            alpha=0.6,
            s=5
        )

    plt.title(f"Dispersão - {lote}")
    plt.xlabel("Tempo (ms)")
    plt.ylabel("ADC")
    plt.legend()
    plt.grid(True)

    plt.show()
```
